```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data =  pd.read_csv("C:\\Users\\Anuruddha\\Desktop\\ML_projects\\18.Movie_Recommendation_System\\database\\movies.csv")
```

## Data preprocessing

```python
# select relaven features

selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
# replace the null values with null strings

for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 selected features

combined_features = movies_data['genres'] +' '+  movies_data['keywords'] +' '+ movies_data['tagline'] +' '+ movies_data['cast'] +' '+ movies_data['director']
```

```python
combined_features
```

```python
# converting text data ---> numerical data

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# Cosine similarity score

similarity  = cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
similarity.shape
```

## Build system

```python
# getting the movie name from the user

movie_name = input("Enter your favorite movie name:")
```

```python
# creating a list of all the movie names given dataset

list_of_al_titles = movies_data['title'].tolist()
```

```python
print(list_of_al_titles)
```

```python
# find the close matchc for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name,list_of_al_titles)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
# find the index of the colst match movie

index_of_the_movie = movies_data[movies_data.title== close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# geting a list of similarity movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
# find higher similarty movie


# sort movies based on their similarity

sorted_similarity_movies = sorted(similarity_score, key= lambda x:x[1], reverse=True)
sorted_similarity_movies
```

```python
# print the name of similar nmoves based on index

print("Movies suggested for you: \n")

i=1

for movie in sorted_similarity_movies:
    index_move = movie[0]
    title_from_index = movies_data[movies_data.index==index_move]['title'].values[0]
    
    # find only 10 movies
    
    if i<11:
        print(i,'.',title_from_index)
        i+=1
```

## Predictive system 

```python
movie_name = input("Enter your favorite movie name:")

list_of_al_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name,list_of_al_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title== close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similarity_movies = sorted(similarity_score, key= lambda x:x[1], reverse=True)


print("Movies suggested for you: \n")

i=1

for movie in sorted_similarity_movies:
    index_move = movie[0]
    title_from_index = movies_data[movies_data.index==index_move]['title'].values[0]
    
    # find only 10 movies
    
    if i<11:
        print(i,'.',title_from_index)
        i+=1
```
